## Import pandas, numpy and folium

```python
import pandas as pd 
import numpy as np
import folium
```

## Import folium map object and display it. We can also zoom in and out and drag the map

```python
world_map = folium.Map()
world_map
```

##  World map centered around India with a 4 zoom level

```python
world_map = folium.Map(location=[20.59, 78.96], zoom_start=4)
world_map 
```

## Stamen Toner Maps for exploring river meanders and coastal zones

```python
world_map = folium.Map(location=[20.59, 78.96], zoom_start=4,tiles='Stamen Toner')
world_map 
```

## Stamen Terrain maps for hill shading and natural vegetation colors

```python
world_map = folium.Map(location=[20.59, 78.96], zoom_start=4,tiles='Stamen Terrain')
world_map 
```
